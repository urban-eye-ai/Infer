```python
import tensorflow as tf

# Create a TFSMLayer from the exported SavedModel.
# Adjust the path if your folder name is different.
tfsmlayer = tf.keras.layers.TFSMLayer("yolov5/runs/train/exp/weights/best_saved_model", call_endpoint='serving_default')

# Build a Keras model using the TFSMLayer.
# Set the input shape to match your model's expected input (here assumed 640x640 RGB image).
input_tensor = tf.keras.Input(shape=(640, 640, 3))
output_tensor = tfsmlayer(input_tensor)
model = tf.keras.Model(inputs=input_tensor, outputs=output_tensor)

# Save the model using the native Keras format.
model.save("best_model.keras")
print("Model exported and saved as best_model.keras")
```
